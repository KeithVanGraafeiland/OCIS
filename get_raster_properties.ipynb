## Describe the properties of rasters in a directory
#### Use get raster properties to describe the raster properties of the rasters in a directory
##### URL: https://pro.arcgis.com/en/pro-app/latest/tool-reference/data-management/get-raster-properties.htm

```python
import arcpy
import os

raster_directory = r"E:\gis_projects\Ocean and Coastal Information System\Raster_Subset\ssp370"

rasters = [os.path.join(raster_directory, f) for f in os.listdir(raster_directory) if f.endswith('.crf')]

def describe_rasters():
    for raster in rasters:
        prop_min = arcpy.management.GetRasterProperties(raster,"MINIMUM")
        prop_max = arcpy.management.GetRasterProperties(raster,"MAXIMUM")
        prop_mean = arcpy.management.GetRasterProperties(raster,"MEAN")
        prop_std = arcpy.management.GetRasterProperties(raster,"STD")
        prop_top = arcpy.management.GetRasterProperties(raster,"TOP")
        prop_left = arcpy.management.GetRasterProperties(raster,"LEFT")
        prop_right = arcpy.management.GetRasterProperties(raster,"RIGHT")
        prop_bottom = arcpy.management.GetRasterProperties(raster,"BOTTOM")
        prop_cellx = arcpy.management.GetRasterProperties(raster,"CELLSIZEX")
        prop_celly = arcpy.management.GetRasterProperties(raster,"CELLSIZEY")
        prop_valuetype = arcpy.management.GetRasterProperties(raster,"VALUETYPE")
        prop_columncount = arcpy.management.GetRasterProperties(raster,"COLUMNCOUNT")
        prop_rowcount = arcpy.management.GetRasterProperties(raster,"ROWCOUNT")
        prop_bandcount = arcpy.management.GetRasterProperties(raster,"BANDCOUNT") 
        prop_anynodata = arcpy.management.GetRasterProperties(raster,"ANYNODATA")
        prop_allnodata = arcpy.management.GetRasterProperties(raster,"ALLNODATA")
        prop_sensorname = arcpy.management.GetRasterProperties(raster,"SENSORNAME") 
        print('Raster Name: ' + os.path.basename(raster))
        print('Raster Path: ' + raster)        
        print(f"MINIMUM:{prop_min}, MAXIMUM:{prop_max}, MEAN:{prop_mean}, STANDARD DEVIATION: {prop_std},\nTOP:{prop_top}, LEFT:{prop_left}, RIGHT:{prop_right}, BOTTOM:{prop_bottom},\nCELLSIZE X:{prop_cellx},CELLSIZE Y:{prop_celly},\nVALUE TYPE:{prop_valuetype}, COLUMN COUNT:{prop_columncount}, ROW COUNT:{prop_rowcount}, BAND COUNT:{prop_bandcount},\nANY NO DATA:{prop_anynodata}, ALL NO DATA:{prop_allnodata}, SENSOR NAME:{prop_sensorname}")
        print('----------------------------------------------------------------------')

```

```python
describe_rasters()
```
