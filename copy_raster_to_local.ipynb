```python
from arcgis.gis import GIS
import arcpy

raster_item_id = 'aa06662b646145a78fb1b471a7bbea25'
gis = GIS("home")
raster_item = gis.content.get(raster_item_id)
item_details = raster_item.get_data()
item_properties = raster_item.properties

print(item_details)
print(item_properties)
output_path = r"E:\gis_projects\hexbins\source_data"
output_name = raster_item.title + ".crf"
output_crf = f"{output_path}/{output_name}"

arcpy.management.CopyRaster(raster_item.url, output_crf, "", "", "", "NONE", "NONE", "", "NONE", "NONE", "CRF")

print(f"Raster downloaded to: {output_crf}")
```
